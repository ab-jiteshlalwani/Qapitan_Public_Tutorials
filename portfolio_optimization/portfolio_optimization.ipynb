# Portfolio Optimization

Easiest example on how to use the Qapitan Platform with the Portfolio Optimization solver.
In this example we send a list of assets and their covariance matrix, as well as the budget we have, and the solver will provide us with the most efficient portfolio with the available solvers in the platform.
This problem can be used as well for asset allocation.

```python
from Qcentroid_SDK import Qcentroid_SDK 
import time
from IPython.display import display, clear_output
```

```python
# Get your private endpoint and credentials from our team - info@qcentroid.xyz
QCENTROID_PUBLIC_API = "https://api.qcentroid.xyz"
PAYLOAD_USER = {"username": "...", "password": "..."}
```

```python
import Parallel
import numpy as np
cryptos=['BTC-USD','ETH-USD','USDT-USD','USDC-USD','BNB-USD','XRP-USD','ADA-USD','SOL-USD','DOGE-USD','DOT-USD','MATIC-USD','DAI-USD','HEX-USD','SHIB-USD','TRX-USD','AVAX-USD','UNI1-USD','ETC-USD','LTC-USD','FTT-USD','ATOM-USD','LINK-USD','CRO-USD','XMR-USD','XLM-USD','BCH-USD','FLOW-USD']
#cryptos=['BTC-USD','ETH-USD','USDT-USD','USDC-USD','BNB-USD']
alpha=0.95
date='2022-09-01'
budget=10000000
def generateJSONLive(assets,date,budget,alpha):
    parquet=Parallel.descargar(assets,None)
    rends=Parallel.rends(parquet,None)
    #rends.replace(np.inf, np.nan)
    rends.replace([np.inf, -np.inf], np.nan, inplace=True)
    cov=Parallel.cov(rends,None)
    ewm=Parallel.EWM(rends,None,alpha)
    data={}
    data["rends"]=list(np.array(ewm.loc[[date]])[0])
    data["current"]=list(np.zeros(len(data['rends'])))
    data["cost"]=list(np.zeros(len(data['rends'])))
    data["cov"]=cov.values.tolist()
    data["budget"]=budget
    data["prices"]=list(np.array(parquet.loc[[date]])[0])
    data["tickers"]=list(ewm.columns)
    d={}
    d["data"]=data
    return d
PAYLOAD = generateJSONLive(cryptos,date,budget,alpha)
```

```python
qcentroid_api = Qcentroid_SDK(QCENTROID_PUBLIC_API, PAYLOAD_USER)
```

```python
header = qcentroid_api.login()
```

```python
print(header)
```

```python
response_json = qcentroid_api.execute(header=header, problem='portfolio-optimization-markovitz', payload=PAYLOAD)
print(response_json)
```

```python
result = qcentroid_api.get_result(header=header, job_name=response_json['job'])
job_id = response_json['job']

while(job_id not in result['job'] or (result['job'][job_id]['status'] != 'FINISHED' and result['job'][job_id]['status'] != 'ERROR')):
	if(job_id in result['job']):
		display(result['job'][job_id]['status'])
	else:
		display('LOADING')
	time.sleep(3)
	clear_output(wait=True)
	result = qcentroid_api.get_result(header=header, job_name=response_json['job'])

clear_output(wait=True)

print("Completed")
qcentroid_api.get_best_result(header, response_json['job'])
```
